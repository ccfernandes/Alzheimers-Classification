```python
# file: Convert_Ecog_Test_Values
# ADNI ecog values in the dataset are separated by question for each cognitive test. 
# For easier input into our model, this script creates an average for each cognitive test per patient.
# example: input --> MEMORY1, MEMORY2...MEMORY8
#          output --> MEMORY_AVG
```

```python
import pandas as pd
import numpy as np
ecogpt = pd.read_csv("C:\\Users\\AZD\\Desktop\\ADNI Dataset\\ECOGPT.csv")
```

```python
data = pd.read_csv("C:\\Users\\AZD\\Desktop\\merged_with_mem_lang_visspat_plan_org.csv")
data = np.append(data,np.zeros([len(data),1]),1)
data
```

```python
# assign a dif value for each cog test 
# df = ecogpt[['MEMORY1','MEMORY2', 'MEMORY3', 'MEMORY4', 'MEMORY5', 'MEMORY6', 'MEMORY7', 'MEMORY8']].copy()
#df = ecogpt[['LANG1','LANG2', 'LANG3', 'LANG4', 'LANG5', 'LANG6', 'LANG7', 'LANG8', 'LANG9']].copy()
#df = ecogpt[['VISSPAT1','VISSPAT2', 'VISSPAT3', 'VISSPAT4', 'VISSPAT6', 'VISSPAT7', 'VISSPAT8']].copy()
df = ecogpt[['DIVATT1','DIVATT2', 'DIVATT3', 'DIVATT4']].copy()

df=df.to_numpy()
```

```python
newCol = 27
i=0
for row in df:
    mysum=0
    counter=4
    for col in row:
        #print(col)
        if col == 0:
            counter=counter-1
        mysum = mysum+col
    mysum=mysum/(counter*4)
    print(i, " ", mysum)
    data[i][newCol]=mysum
    i=i+1
    #avg = np.append(avg, mysum)
    # df1 = pd.DataFrame(df)   
    # df1['Sum'] = mysum
#print(avg)
```
