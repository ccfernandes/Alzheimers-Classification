```python
import pandas as pd
import seaborn as sns
df = pd.read_csv("C:\\Users\\AZD\\Desktop\\merged_normalized_3.csv")
df.head()
```

```python
#histogram plot for ventricles
import matplotlib.pyplot as plt
%matplotlib inline
x1 = df.loc[df.DX_bl=='CN','Ventricles_bl']
x2 = df.loc[df.DX_bl=='MCI','Ventricles_bl']
x3 = df.loc[df.DX_bl=='AD','Ventricles_bl']

kwargs = dict(hist_kws={'alpha':.6},kde_kws={'linewidth':2})
plt.figure(figsize=(15,7),dpi=80)
sns.distplot(x1, color="dodgerblue", label='CN', **kwargs)
sns.distplot(x2, color="green", label='MCI', **kwargs)
sns.distplot(x3, color="deeppink", label='AD', **kwargs)
plt.gca().set(title='Frequency Histogram of baseline Ventricles volume', ylabel='Frequency')
plt.xlim(-2,5)
plt.legend();
```
