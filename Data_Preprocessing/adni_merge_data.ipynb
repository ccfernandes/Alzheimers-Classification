```python
import pandas as pd
import datetime
from functools import reduce
```

```python
# load csv files
ucd_wmh = pd.read_csv("C:\\Users\\AZD\\Desktop\\ADNI Dataset\\ADNI_UCD_WMH_09_01_20.csv")
merge = pd.read_csv("C:\\Users\\AZD\\Desktop\\ADNI Dataset\\ADNIMERGE.csv")
apoeres = pd.read_csv("C:\\Users\\AZD\\Desktop\\ADNI Dataset\\APOERES.csv")
faq = pd.read_csv("C:\\Users\\AZD\\Desktop\\ADNI Dataset\\FAQ.csv")
medhist = pd.read_csv("C:\\Users\\AZD\\Desktop\\ADNI Dataset\\MEDHIST.csv")
ptdemog =  pd.read_csv("C:\\Users\\AZD\\Desktop\\ADNI Dataset\\PTDEMOG.csv")
# dti = pd.read_csv("C:\\Users\\AZD\\Desktop\\ADNI Dataset\\ADNI_DTIROI_10_03_19.csv")
# ecogpt = pd.read_csv("C:\\Users\\AZD\\Desktop\\ADNI Dataset\\ECOGPT.csv")
# ecogsp = pd.read_csv("C:\\Users\\AZD\\Desktop\\ADNI Dataset\\ECOGSP.csv")
```

```python
# select specific columns for each dataframe
sucd_wmh = ucd_wmh[['RID','EXAMDATE','CEREBRUM_TCB', 'TOTAL_HIPPO', 'TOTAL_CSF']].copy()
smerge = merge[['RID','EXAMDATE','AGE', 'APOE4', 'MMSE_bl', 'ADAS11_bl', 'ADAS13_bl', 'Ventricles_bl', 'Entorhinal_bl', 'RAVLT_immediate_bl', 'ABETA_bl', 'DX_bl']].copy()
sapoeres = apoeres[['RID','APGEN1', 'APGEN2']].copy()
sfaq = faq[['RID','USERDATE', 'FAQTOTAL']].copy()
smedhist = medhist[['RID','USERDATE', 'MH2NEURL', 'MH4CARD', 'MH5RESP']].copy()
sptdemog = ptdemog[['RID','USERDATE', 'PTGENDER', 'PTMARRY', 'PTEDUCAT', 'PTRACCAT']].copy()
# secogpt = ecogpt[['RID','MEMORY_TOTAL', 'ORGAN_TOTAL', 'VISSPAT_TOTAL', 'DIVATTEN_TOTAL', 'PLAN_TOTAL']]
```

```python
# select oldest date for each RID per file
sucd_wmh = sucd_wmh.groupby('RID', group_keys=False, sort=False).apply(lambda x:x.sort_values('EXAMDATE')).drop_duplicates('RID', keep='first')
smerge = smerge.groupby('RID', group_keys=False, sort=False).apply(lambda x:x.sort_values('EXAMDATE')).drop_duplicates('RID', keep='first')
# sapoeres = sapoeres.groupby('RID', group_keys=False, sort=False).apply(lambda x:x.sort_values('EXAMDATE')).drop_duplicates('RID', keep='first')
sfaq = sfaq.groupby('RID', group_keys=False, sort=False).apply(lambda x:x.sort_values('USERDATE')).drop_duplicates('RID', keep='first')
smedhist = smedhist.groupby('RID', group_keys=False, sort=False).apply(lambda x:x.sort_values('USERDATE')).drop_duplicates('RID', keep='first')
sptdemog = sptdemog.groupby('RID', group_keys=False, sort=False).apply(lambda x:x.sort_values('USERDATE')).drop_duplicates('RID', keep='first')
```

```python
# rename USERDATE columns so we can differentiate date columns when merged
sfaq=sfaq.rename(columns={'USERDATE':'EXAMDATE_faq'})
smedhist=smedhist.rename(columns={'USERDATE':'EXAMDATE_medhist'})
sptdemog=sptdemog.rename(columns={'USERDATE':'EXAMDATE_ptdemog'})
```

```python
# merge columns one at a time
merge1=pd.merge(smerge, sucd_wmh, left_on="RID", right_on="RID", how="left")
merge1 = merge1.sort_values('RID')
```

```python
merge2 = pd.merge(merge1, sfaq, left_on="RID", right_on="RID", how="left")
merge2
```

```python
merge3 = pd.merge(merge2, smedhist, left_on="RID", right_on="RID", how="left")
merge3
```

```python
merge4 = pd.merge(merge3, sptdemog, left_on="RID", right_on="RID", how="left")
merge4
```
