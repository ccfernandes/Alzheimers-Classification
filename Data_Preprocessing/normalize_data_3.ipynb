```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("C:\\Users\\AZD\\Desktop\\merged_ecog_class.csv")
data
```

```python
col = ["DX_bl", "RID", "EXAMDATE", "AGE", "APOE4", "MH2NEURL", "MH4CARD", "MH5RESP", "PTGENDER", "PTMARRY", "PTEDUCAT", "PTRACCAT"] # remove, not going to normalize

df = data.drop(col, axis=1) #to normalize + abeta
df
```

```python
# # separate ecog columns (already normalized) from columns we want to normalize
df_no_ecog = df.drop(["MEM_AVG", "LANG_AVG", "VISSPAT_AVG", "PLAN_AVG", "ORGAN_AVG", "DIVATT_AVG"], axis=1)
df = df_no_ecog # assign to df 
ecog = data[["MEM_AVG", "LANG_AVG", "VISSPAT_AVG", "PLAN_AVG", "ORGAN_AVG", "DIVATT_AVG"]]

#imaging data 
df_imaging = df.loc[:,["Ventricles_bl", "Entorhinal_bl", "RAVLT_immediate_bl", "CEREBRUM_TCB", "TOTAL_HIPPO", "TOTAL_CSF"]]
#df_imaging
```

```python
#ABETA 
abeta = data["ABETA_bl"] # drop abeta
df = df.drop(["ABETA_bl"], axis=1) #to normalize

for x in range(0,len(abeta)-1): # clean abeta column --> '>1700' cells
    if abeta[x] == '>1700':
        abeta[x] = '1700'
```

```python
df["ABETA_bl"] = abeta
```

```python
# mean normalization for imaging data only
normalized_imaging_df=(df_imaging-df_imaging.mean())/df_imaging.std()

# min-max normalization for the rest
normalized_minmax_df=(df-df.min())/(df.max()-df.min())
```

```python
#normalized_imaging_df
normalized_minmax_df
```

```python
df_col = data[col]
df_col
```

```python
col_ecog = pd.concat([df_col, ecog], axis=1)
col_ecog_abeta = pd.concat([col_ecog, abeta], axis=1)
col_ecog_abeta_imaging = pd.concat([col_ecog_abeta, normalized_imaging_df], axis=1)
merged_normalized_df = pd.concat([col_ecog_abeta_imaging, normalized_minmax_df], axis=1)
```
